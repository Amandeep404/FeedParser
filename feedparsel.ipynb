```python
import feedparser

url = 'https://www.planetpython.org/rss20.xml'

def parse(url):
    return feedparser.parse(url)
    
def get_source(parsed):
    feed = parsed['feed']
    return {
        'link': feed['link'],
        'title': feed['title'],
        'subtitle': feed['subtitle'],
    }

def get_articles(parsed):
    articles = []
    entries = parsed['entries']
    for entry in entries:
        articles.append({
            'id': entry['id'],
            'link': entry['link'],
            'title': entry['title'],
            'summary': entry['summary'],
            'published': entry['published_parsed'],
        })
    return articles
```

```python
url = 'https://feeds.megaphone.fm/strike-force-five'
parsed = parse(url)
get_articles(parsed)
```

```python
aricles= get_articles(parsed)
```
